```r
#Question 2.2
# this script will loop through various lambdas to determine
#highest the highest accuracy. Step increases to
#lambda were set to orders of magnitude of 10

library(kernlab) #load kernlab library
#set parameters ensuring credit card data is set as matrix
x <- as.matrix(Credit_card_data[,1:10])
y <- as.matrix(Credit_card_data[,11])

results <- c(NULL,NULL) #initiating a vector that will log results

#setup loop for a lambda value to understand what magnitude should be used from 10^-4 and 10^7
for(lambda in 10^(-3:7)){
# call ksvm. Vanilladot is a simple linear kernel. 
#model <- ksvm(x,y,scaled=TRUE,type="C-svc",kernel="vanilladot",C=125)
model <- ksvm(x,y,type="C-svc",kernel="vanilladot",C=lambda,scaled=TRUE)
a <- colSums(model@xmatrix[[1]] * model@coef[[1]]) #calculate a1.am
a0 <- model@b # calculate a0
pred <- predict(model,Credit_card_data[,1:10]) # see what the model predicts
# see what fraction of the model's predictions match the actual classification
accuracy <- sum(pred == Credit_card_data[,11]) / nrow(Credit_card_data)
results <- rbind(results,c(lambda,accuracy)) #log the results to determine best accuracy
} #repeat loop
print(results)
#Question 2.2 Answer: A good classifier for the data is when 
# lambda = 10^5 for the kernal vanilladot which classifies the 
# data with 86.39% accuracy with the equation
# -0.08054451 + -0.004117738*V1 + -0.086896089*V2 + 0.129715260*V3 + -0.083744032*V4 + 0.988381368*V5 +
# 0.031253888*V6 + -0.055666972*V7 + -0.037281856*V8 + 0.021940744*V9 + 0.018521785*V10 

## Question 2.3 Try using rbfdot kernel
results_rbf <- c(NULL,NULL) #initiating a vector that will log results

#setup loop for a lambda value to understand what magnitude should be used from 10^-4 and 10^7
for(lambda in 10^c(-3:7)){
  
  # call ksvm. Vanilladot is a simple linear kernel. 
  #model <- ksvm(x,y,scaled=TRUE,type="C-svc",kernel="vanilladot",C=125)
  model <- ksvm(x,y,type="C-svc",kernel="rbfdot",C=lambda,scaled=TRUE)
  a <- colSums(model@xmatrix[[1]] * model@coef[[1]]) #calculate a1.am
  a0 <- model@b # calculate a0
  pred <- predict(model,Credit_card_data[,1:10]) # see what the model predicts
  # see what fraction of the model's predictions match the actual classification
  accuracy <- sum(pred == Credit_card_data[,11]) / nrow(Credit_card_data)
  results_rbf <- rbind(results_rbf,c(lambda,accuracy)) #log the results to determine best accuracy
} #repeat loop
print(results_rbf)
# as lambda approached 10^7, the accuracy increases to 100% meaning
# we are likely beginning to overfit the data


## 2.3 K-nearest neighbor classification
#Source: https://cran.r-project.org/web/packages/kknn/kknn.pdf
# https://www.rdocumentation.org/packages/kknn/versions/1.3.1/topics/train.kknn
library(kknn) #load kknn library

number_of_rows <- nrow(Credit_card_data) ## take the rows of credit card data
Sample <- sample(1:number_of_rows, 2*(number_of_rows/3)) ##take a random sampling of rows between 1 and 654
training<- Credit_card_data[Sample, ] #Taking 2/3 of the data for training
validate<- Credit_card_data[-Sample, ] #taking 1/3 of the data to validate

results_kknn <- c(NULL,NULL,NULL) #initiating a vector that will log results

#kick off loop going through various k values
for(k in (1:15)){
  for(arg in c("rectangular","triangular","epanechnikov", "biweight", "triweight", "cos", "inv", "gaussian","rank")){
credit_kknn <- kknn(V11~., training, validate,k = k, distance = 1, kernel = arg,scale=TRUE)

#summary(credit_kknn)
predicted <- round(fitted(credit_kknn))

#match <- predicted + validate[,11]
#calc_accuracy <- sum(match == 2) / nrow(validate)
calc_accuracy <- sum(predicted == validate[,11]) / nrow(validate)

results_kknn <- rbind(results_kknn,c(arg,k,calc_accuracy))
}}
print(results_kknn)

#Answer: Using the kernel "inv", a K value of 13 is a good value.
# This classifies the data points in the set with ~87.61% accuracy
```
